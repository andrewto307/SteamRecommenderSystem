```python
import ast
import pandas as pd
import json

records = []
with open("australian_user_reviews.json", "rt", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        obj = ast.literal_eval(line)  
        records.append(obj)

df1 = pd.DataFrame(records)

print(df1.shape)
print(df1.head())
```

```python
import ast
import pandas as pd
import json


records = []
with open("australian_users_items.json", "rt", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        obj = ast.literal_eval(line) 
        records.append(obj)

df2 = pd.DataFrame(records)

print(df2.shape)
print(df2.head())
```

## DF1: User Review

```python
df1 = df1.explode('reviews').reset_index(drop=True)
```

```python
def parse_review(x):
    if isinstance(x, dict):
        review_text = x.get('review', '')
        posted_date = x.get('posted', '')
        item_id = x.get('item_id', '')
        recommend = x.get('recommend', '')

        return {
            'review_text': review_text,
            'posted': posted_date,
            'item_id': item_id,
            'recommend': recommend
        }
    
    return {"review_text": "", "posted": "", "item_id": "", "recommend": ""}



df1 = df1.join(df1['reviews'].apply(parse_review).apply(pd.Series))
```

```python
df1.drop(columns=['reviews'], inplace=True)
```

```python
from datetime import datetime

df1['posted'] = df1['posted'].astype(str).str.replace(r'Posted |[.]', '', regex=True)
df1['posted'] = pd.to_datetime(df1['posted'], errors='coerce')
df1
```

```python
df1['user_id'] = df1['user_id'].astype('string')
df1['user_url']= df1['user_url'].astype('string')
df1['review_text'] = df1['review_text'].astype('string')
df1['item_id'] = pd.to_numeric(df1['item_id'], errors='coerce').dropna()
df1['recommend'] = df1['recommend'].astype(bool)
```

## DF2: User-Item


```python
df2_test = df2.copy()
```

```python
df2_test = df2_test.explode('items').reset_index(drop=True)
```

```python
df2_test.drop(columns=['user_url', 'steam_id', 'items_count'], inplace=True)
df1.drop(columns='user_url', inplace=True)
```

```python
df1
```

```python
df1_df2 = pd.merge(df1, df2_test, on=['user_id'], how='inner')
df1_df2
```

```python
df1_df2.dropna(subset=['items', 'item_id'], inplace=True)
```

```python
def parse_item(x):
    if isinstance(x, dict):
        item_id = x.get('item_id', '')
        item_name = x.get('item_name', '')
        playtime_forever = x.get('playtime_forever', '')
        playtime_2weeks = x.get('playtime_2weeks', '')

        return {
            'item_id': item_id,
            'item_name': item_name,
            'playtime_forever': playtime_forever,
            'playtime_2weeks': playtime_2weeks
        }
    
    return {"item_id": "", "item_name": "", "playtime_forever": "", "playtime_2weeks": ""}



df1_df2 = df1_df2.join(df1_df2['items'].apply(parse_item).apply(pd.Series).add_suffix('_parsed'))
```

```python
df1_df2
```

```python
df1_df2.dropna(subset=['item_id_parsed', 'item_id'], inplace=True)
```

```python
df1_df2['item_id_parsed'] = pd.to_numeric(df1_df2['item_id_parsed'], errors='coerce').dropna()
```

```python
df1_df2 = df1_df2.loc[df1_df2['item_id'] == df1_df2['item_id_parsed']]
```

```python
df1_df2.info()
```

```python
df1_df2.reset_index(drop=True, inplace=True)
```

```python
df1_df2.drop(columns=['items', 'item_id_parsed'], inplace=True)
df1_df2.rename(columns={'item_name_parsed': 'item_name', 'playtime_forever_parsed': 'playtime_forever', 'playtime_2weeks_parsed': 'playtime_2weeks'}, inplace=True)
df1_df2
```
